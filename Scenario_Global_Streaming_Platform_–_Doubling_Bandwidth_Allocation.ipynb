```python
# # Scenario: Global Streaming Platform – Doubling Bandwidth Allocation
# # The Problem Statement
# # Every day, your platform receives millions of video stream logs from users worldwide.
#  Due to a new regulation, you must double the bandwidth allocation for every stream to ensure smooth playback.
#   Finally, you need a consolidated list of updated bandwidth values to calculate infrastructure costs.

# # The Challenge
# # Data Size: Millions of logs are too large for a single machine.

# # Speed: You need results instantly to adjust server capacity.
```

```python
# ==============================
# 1. Spark Session Setup
# ==============================
from pyspark.sql import SparkSession
from pyspark.sql.functions import col

spark = SparkSession.builder \
    .appName("GlobalStreaming_Bandwidth_Update") \
    .getOrCreate()


# ==============================
# 2. Read Input Data (logs.csv)
# ==============================
df = spark.read.csv("log.csv", header=True, inferSchema=True)

print("=== Original Data ===")
df.show()


# ==============================
# 3. Core Transformation
# Double the bandwidth
# ==============================
updated_df = df.withColumn(
    "updated_bandwidth",
    col("bandwidth") * 2
)

print("=== Updated Data ===")
updated_df.show()


# ==============================
# 4. Optimization (Optional but good practice)
# ==============================
updated_df = updated_df.repartition(4)


# ==============================
# 5. Select Required Output
# ==============================
final_df = updated_df.select("stream_id", "updated_bandwidth")


# ==============================
# 6. Save Output
# ==============================
final_df.write \
    .mode("overwrite") \
    .csv("output_logs")

print("✅ Processing Complete. Output saved in 'output_logs' folder.")


# ==============================
# 7. Stop Spark Session
# ==============================
spark.stop()
```
