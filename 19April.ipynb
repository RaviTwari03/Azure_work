```python
import requests
from google.colab import userdata

api_key = userdata.get('Ai-foundry')

if not api_key:
    raise ValueError(" API key not found.")

url = "https://harshita-ai-foundry.services.ai.azure.com/api/projects/proj-default/applications/email-assistant/protocols/openai/responses?api-version=2025-11-15-preview"

headers = {
    "Content-Type": "application/json",
    "api-key": api_key
}

print(" Email Assistant is running! Type 'exit' to stop.\n")

while True:
    user_input = input("Enter your email request: ")

    if user_input.lower() in ["exit", "quit"]:
        print("Exiting... ")
        break

    data = {
        "model": "gpt-4.1",
        "input": user_input
    }

    response = requests.post(url, headers=headers, json=data)
    result = response.json()

    if "output" in result:
        reply = result["output"][0]["content"][0]["text"]

        print("\n Generated Email:\n")
        print(reply)
        print("\n" + "-"*50 + "\n")
    else:
        print(" API Error Response:\n", result)
```

```python
import requests
from google.colab import userdata

# API Key
api_key = userdata.get('Ai-foundry')

# Replace with your endpoint
url = "https://harshita-ai-foundry.services.ai.azure.com/api/projects/proj-default/applications/interview-simulator/protocols/openai/responses?api-version=2025-11-15-preview"

headers = {
    "Content-Type": "application/json",
    "api-key": api_key
}

print(" AI Interview Simulator Started!\n")

mode = input("Choose mode (HR / Technical / Stress): ")

while True:
    user_input = input("\nYour Answer (or type 'exit'): ")

    if user_input.lower() == "exit":
        print("Interview ended. Good luck! ")
        break

    prompt = f"""
Mode: {mode}

User Answer:
{user_input}

Now:
1. Ask next question
2. Evaluate this answer with:
   - Score (out of 10)
   - Strengths
   - Weaknesses
   - Improved answer
   - Tips
"""

    data = {
        "model": "gpt-4.1",
        "input": prompt
    }

    response = requests.post(url, headers=headers, json=data)
    result = response.json()

    if "output" in result:
        reply = result["output"][0]["content"][0]["text"]
        print("\nAI:\n", reply)
    else:
        print("Error:", result)
```
